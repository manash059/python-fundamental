simple Linear Regression Model

```python
ml_v=smf.ols('MPG~VOL',data = cars).fit()
ml_v.rsquared
```

```python
ml_v=smf.ols('MPG~WT',data = cars).fit()
ml_v.rsquared
```

```python
ml_v=smf.ols('MPG~HP',data = cars).fit()
ml_v.rsquared
```

```python
ml_v.params
```

```python
ml_w=smf.ols('MPG~VOL',data = cars).fit()
ml_w.rsquared
```

```python
ml_sp=smf.ols('MPG~SP',data = cars).fit()
ml_sp.rsquared
```

```python
import statsmodels.formula.api as smf
ml_v=smf.ols('MPG~VOL+SP+HP+WT',data = cars).fit()
ml_v.rsquared
```
